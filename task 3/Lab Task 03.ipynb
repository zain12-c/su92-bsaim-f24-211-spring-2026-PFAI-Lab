```python
# Water Jug Problem using DFS
capacity_a = 4
capacity_b = 3
target = 2

visited_states = set()

def depth_first_search(jug_a, jug_b, steps):

    if (jug_a, jug_b) in visited_states:
        return False

    visited_states.add((jug_a, jug_b))

    steps.append((jug_a, jug_b))

    if jug_a == target:
        print("Solution Found\n")
        print("Steps:")
        for s in steps:
            print(s)
        return True

    if jug_a < capacity_a:
        print("Rule: Fill Jug A")
        if depth_first_search(capacity_a, jug_b, steps.copy()):
            return True

    if jug_b < capacity_b:
        print("Rule: Fill Jug B")
        if depth_first_search(jug_a, capacity_b, steps.copy()):
            return True

    if jug_a > 0:
        print("Rule: Empty Jug A")
        if depth_first_search(0, jug_b, steps.copy()):
            return True

    if jug_b > 0:
        print("Rule: Empty Jug B")
        if depth_first_search(jug_a, 0, steps.copy()):
            return True

    if jug_a > 0 and jug_b < capacity_b:
        move = min(jug_a, capacity_b - jug_b)
        print("Rule: Pour Jug A -> Jug B")
        if depth_first_search(jug_a - move, jug_b + move, steps.copy()):
            return True

    if jug_b > 0 and jug_a < capacity_a:
        move = min(jug_b, capacity_a - jug_a)
        print("Rule: Pour Jug B -> Jug A")
        if depth_first_search(jug_a + move, jug_b - move, steps.copy()):
            return True

    return False

depth_first_search(0, 0, [])
```
