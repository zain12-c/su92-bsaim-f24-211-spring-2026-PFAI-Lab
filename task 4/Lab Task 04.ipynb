```python
def check_position(chessboard, r, c, size):
    # Check column conflicts
    for x in range(r):
        if chessboard[x][c] == 1:
            return False

    # Check upper-left diagonal
    x, y = r - 1, c - 1
    while x >= 0 and y >= 0:
        if chessboard[x][y] == 1:
            return False
        x -= 1
        y -= 1

    # Check upper-right diagonal
    x, y = r - 1, c + 1
    while x >= 0 and y < size:
        if chessboard[x][y] == 1:
            return False
        x -= 1
        y += 1

    return True


def place_queens(chessboard, r, size, result_list):
    if r == size:
        # Found a solution
        result_list.append(["".join("Q" if cell == 1 else "." for cell in row) for row in chessboard])
        return

    for c in range(size):
        if check_position(chessboard, r, c, size):
            chessboard[r][c] = 1
            place_queens(chessboard, r + 1, size, result_list)
            chessboard[r][c] = 0   # Backtrack


def n_queen_solver(size):
    chessboard = [[0 for _ in range(size)] for _ in range(size)]
    result_list = []
    place_queens(chessboard, 0, size, result_list)
    return result_list


# Example usage
board_size = 8
answers = n_queen_solver(board_size)

print(f"Total solutions for {board_size}-Queens: {len(answers)}\n")

for solution in answers:
    for line in solution:
        print(line)
    print()
```
